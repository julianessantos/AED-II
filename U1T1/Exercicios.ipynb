### *Questão 2.1:*
Suppose you’re tossing two coins at the same time. They’re loaded
in different ways, according to the table below. Calculate the
probability of getting all possible outcomes:

| p1(H) | p2(H) | H-H | H-T | T-H | T-T |
|-------|-------|-----|-----|-----|-----|
| 0.5   | 0.5   |     |     |     |     |
| 0.6   | 0.7   |     |     |     |     |
| 0.4   | 0.8   |     |     |     |     |
| 0.1   | 0.2   |     |     |     |     |
| 0.3   | 0.4   |     |     |     |     |

*Resposta:*

Para calcular a chance de dois eventos independentes acontecerem ao mesmo tempo (como o resultado de cada moeda), a gente multiplica as chances de cada um acontecer.
Dessa forma, a probabilidade de ambas as moedas caírem em cara (H-H) será de:

<center><p>P(H-H) = P(H) * P(H)</p></center>

A probabilidade de uma moeda cair cara e a outra coroa (H-T) será:

<center><p>P(H-T) = P(H) * P(T)</p></center>

E assim consecutivamente, com isso a tabela ficará:

| p1(H) | p2(H) | H-H | H-T | T-H | T-T |
|-------|-------|-----|-----|-----|-----|
| 0.5   | 0.5   | 0.25| 0.25| 0.25| 0.25|
| 0.6   | 0.7   | 0.42| 0.18| 0.28| 0.12|
| 0.4   | 0.8   | 0.32| 0.08| 0.48| 0.12|
| 0.1   | 0.2   | 0.02| 0.08| 0.18| 0.72|
| 0.3   | 0.4   | 0.12| 0.18| 0.28| 0.42|

### *Questão 2.2:*
60% of the emails hitting my inbox is spam. You design a phenomenal spam filter which is able to tell me, with 98% accuracy, whether an email is spam or not: if an email is not spam, the system has a 98% probability of saying so. The filter knows 60% of emails are spam and so it will flag 60% of my emails. Suppose that, at the end of the week, I look in my spam box and see 963 emails. Use Bayes' Theorem to calculate how many of those 963 emails in my spam box I should suspect to be non-spam.

*Resposta:*

1. Definir os Eventos:

  S: O email é spam.
  
  N: O email não é spam (é legítimo).

  F: O filtro classifica o email como spam.

2. Estabelecer as Probabilidades:

  P(S) = 0.6 (60% dos emails são spam)
  
  P(F|S) = 0.98 (98% de chance de o filtro classificar um spam como spam)
  
  P(N|N) = 0.98 (98% de chance de o filtro classificar um email legítimo como legítimo)

3. Calculando:

Com o auxilio do Teorema de Bayes podemos calcular a probabilidade de um email legítimo não ser spam

                                 P(N|F) = P(F|N) * P(N) / P(F)

  Para calcularmos P(F)

  <center><p>P(F) = P(F|S) * P(S) + P(F|N) * P(N)</p></center>

  Onde

<center><p>P(F|N) = 1 - P(N|F) = 1 - 0.98 </p></center>

<center><p>P(N) = 1 - P(S) = 1 - 0.8 = 0.4</p></center>

<center><p>P(N) = 1 - P(S) = 1 - 0.8 = 0.4</p></center>

Logo

                             P(F) = 0.98 * 0.6 + 0.02 * 0.4 = 0.596

Sendo assim:

                              P(N|F) = 0.02 * 0.4 / 0.596 ≈ 0.0134

Dessa forma, o número de emails legitimos marcados como spam foi:

                               P(N|F) * 963 ≈ 0.0134 * 963 ≈ 12.9

### *Questão 2.4:*
Draw the probability mass function and the cumulative distribution of the following outcome probabilities:
	p

|Outcome | p |
|---|------|
| 1 | 0.1  |
| 2 | 0.15 |
| 3 | 0.2  |
| 4 | 0.21 |
| 5 | 0.17 |
| 6 | 0.09 |
| 7 | 0.06 |
| 8 | 0. 2|


*Resposta:*

Função de Massa de Probabilidade (PMF): O gráfico de barras mostra a probabilidade de cada resultado individual.

Função de Distribuição Acumulada (CDF): O gráfico de linha mostra a probabilidade acumulada até cada resultado, variando de 0 a 1.


```python
import matplotlib.pyplot as plt
import numpy as np

# Dados
resultados = np.array([1, 2, 3, 4, 5, 6, 7, 8])
probabilidades = np.array([0.1, 0.15, 0.2, 0.21, 0.17, 0.09, 0.06, 0.02])

# Função de Massa de Probabilidade (PMF)
plt.figure(figsize=(5, 2.5))
plt.bar(resultados, probabilidades, color='b', alpha=0.6)
plt.title('Função de Massa de Probabilidade (PMF)', fontsize=14)
plt.xlabel('Resultado', fontsize=12)
plt.ylabel('Probabilidade', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(resultados)
plt.show()

# Cálculo da Função de Distribuição Acumulada (CDF)
cdf = np.cumsum(probabilidades)

# Função de Distribuição Acumulada (CDF)
plt.figure(figsize=(5, 2.5))
plt.step(resultados, cdf, where='mid', color='r', linewidth=2, alpha=0.8)
plt.fill_between(resultados, cdf, step="mid", alpha=0.3, color='r')
plt.title('Função de Distribuição Acumulada (CDF)', fontsize=14)
plt.xlabel('Resultado', fontsize=12)
plt.ylabel('Probabilidade Acumulada', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(resultados)
plt.ylim(0, 1.05)
plt.show()
```

### *Questão 2.5:*
How many bits do we need to independently encode v1 and v2 from this data? How much would we save in encoding v1 if we knew v2?

*Resposta:*

A quantidade de bits necessária para codificar v1 e v2 independentemente depende de suas entropias individuais,
𝐻(𝑣1) e 𝐻(𝑣2). O número de bits necessário é a soma dessas entropias:
𝐻(𝑣1)+𝐻(𝑣2).

Se soubéssemos v2, poderíamos usar a entropia condicional 𝐻(𝑣1∣𝑣2), que representa a incerteza restante em v1 dado v2. A economia seria a diferença
𝐻(𝑣1)−𝐻(𝑣1∣𝑣2).
